```python
import json
from collections import Counter

# Load JSON 
with open("dataset/dataset_query_decision_gt.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Count categories
category_counts = Counter(item["category"] for item in data["dataset"])

print("number of queries:", len(data["dataset"]))
print("\nCategory distribution:")
for category, count in category_counts.items():
    print(f"{category}: {count}")


print("\nPercentage:")
total = len(data["dataset"])
for category, count in category_counts.items():
    print(f"{category}: {100*count/total:.2f}%")
```
